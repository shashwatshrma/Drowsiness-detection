```python
import cv2
from tensorflow.keras.models import load_model
import numpy as np
from pygame import mixer
import time
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_righteye_2splits.xml')
model = load_model('./model checkpoint/model.keras')
mixer.init()
sound = mixer.Sound('./alarm.wav')
score = 0
```

```python
cap = cv2.VideoCapture(0)
# uncomment this and the statements below to save the eye image sent to the model along with the prediction value in saved/ folder
# i = 0
while True:
    time.sleep(1)
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if ret:
        height, width = frame.shape[0:2]
        faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.2, minNeighbors = 3)
        cv2.rectangle(frame, (0, height-50), (200, height), (0, 0, 0), thickness = cv2.FILLED)
        for (x,y,w,h) in faces:
            cv2.rectangle(frame, pt1 = (x, y), pt2 = (x+w, y+h), color = (255, 0, 0), thickness = 3)
            face_frame = gray[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(face_frame, scaleFactor = 1.3, minNeighbors = 3)
            if(len(eyes) == 0): 
                continue
            (ex,ey,ew,eh) = eyes[0]
            cv2.rectangle(frame, pt1=(x+ex,y+ey), pt2=(x+ex+ew, y+ey+eh), color = (255, 0, 0), thickness = 3)
            eye = gray[y+ey:y+ey+eh, x+ex:x+ex+ew]
            # oeye=eye
            eye = cv2.resize(eye, (80, 80))
            eye = np.expand_dims(eye, axis=-1) # need to make the image (80, 80, 1)
            eye = np.repeat(eye, 3, axis=2) # need to make the image (80, 80, 3)
            eye = eye/255
            # i+=1;
            eye = np.expand_dims(eye, axis=0) # need to make the image (1, 80, 80, 3) as that is passed to the model
            ## preprocessing is done now model predict
            prediction = model.predict(eye)
            # cv2.imwrite('saved/'+str(i)+'_'+str(prediction[0][0])+'.jpg', oeye)
            if prediction[0][0] < 0.75 :
                cv2.putText(frame,'closed',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(255,255,255),thickness=1)
                cv2.putText(frame,'score'+str(score),(100,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(255,255,255),thickness=1)
                score = score+1
                if score > 3:
                    try:
                        sound.play()
                    except:
                        pass
    
            else:
                cv2.putText(frame,'open',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(255,255,255),thickness=1)
                cv2.putText(frame,'score'+str(score),(100,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(255,255,255),thickness=1)
                score = 0
            
        cv2.imshow('frame', frame) 
    if cv2.waitKey(33) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
